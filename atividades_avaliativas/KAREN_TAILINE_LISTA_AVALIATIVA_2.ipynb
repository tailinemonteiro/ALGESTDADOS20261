```python
# 1) Levantamento Salarial

# variáveis
soma_salarios = 0
maior_salario = 0
contador_acima_3000 = 0

# repetir 10 vezes (um para cada funcionário)
for i in range(1, 11):

    # pedindo o salário do funcionário
    salario = float(input(f"Digite o salário do {i}° funcionário: "))

    # somando o salário na variável total
    soma_salarios += salario

    # se o salário é maior que o maior já registrado
    if salario > maior_salario:
        maior_salario = salario

    # verificando se o salário é maior que 3000
    if salario > 3000:
        contador_acima_3000 += 1

# calculando a média salarial
media = soma_salarios / 10

# exibe resultados
print("\nRESULTADOS:")
print("Média salarial:", media)
print("Maior salário:", maior_salario)
print("Funcionários que ganham acima de R$ 3000:", contador_acima_3000)
```

```python
# 2) Desempenho da Turma

# variáveis
soma_notas = 0
total_alunos = 0
aprovados = 0
reprovados = 0

# só vai parar quando o usuário digitar -1
while True:

     # pedindo a nota do aluno
    nota = float(input("Digite a nota do aluno (-1 para encerrar): "))

    # verificando se o usuário quer encerrar
    if nota == -1:
        break  # para sair do loop

    # somando a nota
    soma_notas += nota

    # contando mais um aluno
    total_alunos += 1

    # verificar se foi aprovado ou reprovado
    if nota >= 7:
        aprovados += 1
    else:
        reprovados += 1

# verificar se pelo menos um aluno foi inserido
if total_alunos > 0:

    # calculando a média
    media = soma_notas / total_alunos

    print("\nRESULTADOS:")
    print("Média da turma:", media)
    print("Aprovados:", aprovados)
    print("Reprovados:", reprovados)

else:
    print("Nenhuma nota foi informada.")
```

```python
# 3) Sistema Bancário de Saques

# pedindo o saldo inicial
saldo = float(input("Digite o saldo inicial da conta: "))

# loop para permitir vários saques
while True:

    # pedindo o valor do saque
    saque = float(input("Digite o valor do saque (0 para encerrar): "))

    # verificar se o usuário quer encerrar
    if saque == 0:
        print("Encerrando operações...")
        break

    # verificar se há saldo suficiente
    if saque <= saldo:
        # atualizar o saldo após o saque
        saldo -= saque
        print("Saque realizado com sucesso!")
        print("Saldo atual:", saldo)
    else:
        # caso não tenha saldo suficiente
        print("Saldo insuficiente! Tente outro valor.")

# mostrando o saldo final
print("Saldo final da conta:", saldo)
```

```python
# 4) Pesquisa Demográfica

# variáveis
soma_idades = 0
total = 0
maior_idade = 0
maiores_18 = 0

# loop para ler várias idades
while True:
    idade = int(input("Digite a idade (0 para encerrar): "))

    # condição de parada
    if idade == 0:
        break

    # soma das idades
    soma_idades += idade

    # conta quantas pessoas foram informadas
    total += 1

    # verifica maior idade
    if idade > maior_idade:
        maior_idade = idade

    # conta maiores de 18 anos
    if idade > 18:
        maiores_18 += 1

# verifica se houve dados
if total > 0:
    # calcula média
    media = soma_idades / total

    # exibe resultados
    print("Média:", media)
    print("Maior idade:", maior_idade)
    print("Maiores de 18:", maiores_18)
```

```python
# 5) Controle de Acesso

# define a senha correta
senha_correta = 2024

# contador de tentativas
tentativas = 0

# permite até 3 tentativas
while tentativas < 3:

    # solicita senha
    senha = int(input("Digite a senha (Apenas números são aceitos): "))

    # verifica se está correta
    if senha == senha_correta:
        print("Acesso liberado!")
        break
    else:
        print("Senha incorreta!")
        tentativas += 1  # soma tentativa

# se exceder 3 tentativas
if tentativas == 3:
    print("Acesso bloqueado!")
```

```python
# 6) Vendas com Desconto

# variável acumuladora do total
total = 0

# loop para 8 vendas
for i in range(1, 9):

    # entrada do valor
    valor = float(input(f"Digite o valor da {i}° venda: "))

    # verifica se tem desconto
    if valor > 100:
        valor = valor * 0.9  # aplica 10%

    # soma no total
    total += valor

# exibe resultado final
print("Total arrecadado:", total)
```

```python
# 7) Análise de Sequência Numérica

# inicialização dos contadores
pares = 0
impares = 0
soma_positivos = 0

# loop até digitar 0
while True:
    num = int(input("Digite um número (0 para encerrar): "))

    # parada
    if num == 0:
        break

    # verifica par ou ímpar
    if num % 2 == 0:
        pares += 1
    else:
        impares += 1

    # soma apenas positivos
    if num > 0:
        soma_positivos += num

# mostra resultados
print("Pares:", pares)
print("Ímpares:", impares)
print("Soma dos positivos:", soma_positivos)
```

```python
# 8) Monitoramento de Peso

# variáveis
soma = 0
acima_80 = 0

# ler o primeiro peso separadamente
peso = float(input("Digite o 1° peso: "))

# considera o primeiro peso como o menor inicialmente
menor_peso = peso

# soma o primeiro peso
soma += peso

# verifica se é maior que 80
if peso > 80:
    acima_80 += 1

# for para os outros 5 pesos
for i in range(2, 7):

    # pede os próximos pesos
    peso = float(input(f"Digite o {i}° peso: "))

    # soma dos pesos
    soma += peso

    # verifica se é maior que 80
    if peso > 80:
        acima_80 += 1

    # verifica o menor peso
    if peso < menor_peso:
        menor_peso = peso

# calcula a média
media = soma / 6

# mostra os resultados
print("Média dos pesos:", media)
print("Quantidade acima de 80kg:", acima_80)
print("Menor peso:", menor_peso)
```

```python
# 9) Votação Eletrônica

# variáveis de votos
c1 = 0
c2 = 0
c3 = 0

# loop de votação
while True:

    # entrada do voto
    voto = int(input("Vote (1,2,3 ou 0 para sair): "))

    # encerramento
    if voto == 0:
        break

    # contagem de votos
    if voto == 1:
        c1 += 1
    elif voto == 2:
        c2 += 1
    elif voto == 3:
        c3 += 1

# mostra votos
print("C1:", c1)
print("C2:", c2)
print("C3:", c3)

# verifica vencedor
if c1 > c2 and c1 > c3:
    print("Vencedor: 1")
elif c2 > c1 and c2 > c3:
    print("Vencedor: 2")
else:
    print("Vencedor: 3")
```

```python
# 10) Produção Semanal

# variáveis
soma = 0
maior = 0
dias_baixos = 0

# loop de 7 dias
for i in range(1, 8):

    # entrada
    prod = int(input(f"Produção {i}° dia: "))

    # soma produção
    soma += prod

    # maior valor
    if prod > maior:
        maior = prod

    # conta dias abaixo de 100
    if prod < 100:
        dias_baixos += 1

# média
media = soma / 7

# resultados
print("Média:", media)
print("Maior:", maior)
print("Dias abaixo de 100:", dias_baixos)
```

```python
# 11) Múltiplos de Três

# variáveis
soma = 0
contador = 0

# loop até -1
while True:

    # entrada
    num = int(input("Digite um número (-1 para sair): "))

    # parada
    if num == -1:
        break

    # múltiplo de 3
    if num % 3 == 0:
        soma += num
        contador += 1

# calcula média
if contador > 0:
    media = soma / contador
    print("Quantidade:", contador)
    print("Média:", media)
```

```python
# 12) Altura dos Alunos

# variáveis
soma = 0

# as 5 alturas
for i in range(1, 6):

    # ler a altura
    altura = float(input(f"Digite a altura do {i}° aluno: "))

    # somando as alturas
    soma += altura

# calculando a média
media = soma / 5

# contador de alunos acima da média
acima = 0

print("\nDigite novamente a altura dos alunos para verificar média e acima da média: ")
# ler novamente as 5 alturas
for i in range(1, 6):
    altura = float(input(f"Digite novamente a altura do {i}° aluno: "))

    # verificar se está acima da média
    if altura > media:
        acima += 1

# mostrando resultados
print("\nMédia das alturas:", media)
print("Quantidade acima da média:", acima)
```

```python
# 13) Registro de Depósitos

# variáveis
total = 0
maior = 0
quantidade = 0

# loop até 0
while True:
    valor = float(input("Digite o valor do depósito (0 para sair): "))

    # parada
    if valor == 0:
        break

    # soma total
    total += valor

    # conta depósitos
    quantidade += 1

    # verifica maior valor
    if valor > maior:
        maior = valor

# resultado
print("Total:", total)
print("Maior:", maior)
print("Quantidade:", quantidade)
```

```python
# 14) Verificador de Número Primo

# entrada do número
num = int(input("Digite um número: "))

# contador de divisores
divisores = 0

# verificar divisibilidade
for i in range(1, num + 1):

    # se dividir exatamente
    if num % i == 0:
        print("Divisor:", i)
        divisores += 1

# mostra quantidade
print("Total divisores:", divisores)

# verifica se é primo
if divisores == 2:
    print("É primo")
else:
    print("Não é primo")
```

```python
# 15) Loja de Eletrônicos

# variável
total = 0

# loop até digitar 0
while True:

    # entrada
    valor = float(input("Digite o preço (0 para sair): "))

    # parada
    if valor == 0:
        break

    # aplica desconto se necessário
    if valor > 200:
        valor = valor * 0.95

    # soma ao total
    total += valor

# resultado final
print("Total a pagar:", total)
```

```python
# 16) Tempo de Atendimento

# pede o tempo do primeiro atendimento
tempo = float(input("Digite o tempo de atendimento (0 para encerrar): "))

# variáveis
soma = 0
contador = 0

# enquanto não digitar 0, continua
while tempo != 0:
    soma += tempo
    contador += 1

    # pede o próximo tempo
    tempo = float(input("Digite o tempo de atendimento (0 para encerrar): "))

# quando sair do loop:
if contador > 0:
    media = soma / contador

    print(f"Total de clientes: {contador}")
    print(f"Tempo médio: {media:.2f}")

else:
    print("Ninguém foi atendido")
```

```python
# 17) Análise de 10 Números

# variáveis
negativos = 0
soma_positivos = 0

# vai ler os 10 números
for i in range(1, 11):
    numero = int(input(f"Digite o {i}° número: "))

    if i == 1:
        maior = numero
    else:
        if numero > maior: # compara pra ver se é o maior
            maior = numero

    # verifica se é negativo
    if numero < 0:
        negativos += 1

    # verifica se é positivo
    if numero > 0:
        soma_positivos += numero

# depois que termina todas as verificações:
print(f"\nQuantidade de negativos: {negativos}")
print(f"Soma dos positivos: {soma_positivos}")
print(f"Maior valor: {maior}")
```

```python
# 18) Consumo de Energia

# variável
total = 0

# vai ler o consumo das 6 casas
for i in range(1, 7):
    consumo = float(input(f"Digite o consumo da {i}° casa (em kWh): "))

    # se passar de 100, é mais caro
    if consumo > 100:
        valor = consumo * 0.80
    else:
        valor = consumo * 0.50

    total += valor

# depois de tudo, mostra quanto arrecadou
print(f"\nTotal arrecadado: R$ {total:.2f}")
```

```python
# 20) Estatísticas de Notas

# pede a primeira nota
nota = float(input("Digite a nota (-1 para encerrar): "))

# variáveis
soma = 0
contador = 0

# enquanto não digitar -1
while nota != -1:
    soma += nota
    contador += 1

    # na primeira nota, define como maior
    if contador == 1:
        maior = nota
    else:
        # verifica se é a maior
        if nota > maior:
            maior = nota

    # pede a próxima nota
    nota = float(input("Digite a nota (-1 para encerrar): "))

# depois que termina
if contador > 0:
    media = soma / contador   # para calcular a média
    abaixo_media = 0
    print(f"\nMédia das notas: {media:.2f}")
    print(f"Maior nota: {maior}")

    # pede as notas novamente
    print("\nDigite as notas novamente para verificar quais estão abaixo da média: ")

    for i in range(1, contador + 1):
        n = float(input(f"Digite a {i}° nota: "))

        if n < media:
            abaixo_media += 1

    # mostra quantas ficaram abaixo da média
    print(f"\nNotas abaixo da média: {abaixo_media}")

else:
    print("Nenhuma nota foi digitada")
```

```python
# 21) Pesquisa Eleitoral

# variáveis
votos1 = 0
votos2 = 0
votos3 = 0
invalidos = 0

# pede o primeiro voto
voto = int(input("Digite o voto (1, 2, 3 ou 0 para encerrar): "))

# enquanto não for 0
while voto != 0:
    if voto == 1:
        votos1 += 1
    elif voto == 2:
        votos2 += 1
    elif voto == 3:
        votos3 += 1
    else:
        invalidos += 1   # qualquer outro valor é inválido

    # pede o próximo voto
    voto = int(input("Digite o voto (1, 2, 3 ou 0 para encerrar): "))

# calcula os válidos
validos = votos1 + votos2 + votos3

# descobre o vencedor
if votos1 > votos2 and votos1 > votos3:
    vencedor = "Candidato 1"
elif votos2 > votos1 and votos2 > votos3:
    vencedor = "Candidato 2"
elif votos3 > votos1 and votos3 > votos2:
    vencedor = "Candidato 3"
else:
    vencedor = "Empate"

# mostra o resultado
print(f"Votos válidos: {validos}")
print(f"Votos inválidos: {invalidos}")
print(f"Vencedor: {vencedor}")
```

```python
# 22) Estacionamento

# variável
total = 0

# pede o tempo do primeiro carro
tempo = float(input("Digite o tempo em horas (0 para encerrar): "))

# enquanto não digitar 0, continua
while tempo != 0:

    # calcula o valor normal
    valor = tempo * 10

    #passou de 5 horas, dá desconto
    if tempo > 5:
        valor = valor * 0.9

    total += valor

    # pede o próximo tempo
    tempo = float(input("Digite o tempo (horas) (0 para encerrar): "))

# mostra o total
print(f"\nTotal arrecadado: R$ {total:.2f}")
```

```python
# 24) Classificação Escolar

# variáveis
aprovados = 0
recuperacao = 0
reprovados = 0

# vai ler as 8 médias
for i in range(1, 9):
    media = float(input(f"Digite a média do {i}° aluno: "))

    # classifica a média
    if media >= 7:
        print("Aprovado")
        aprovados += 1

    elif media >= 5:
        print("Recuperação")
        recuperacao += 1

    else:
        print("Reprovado")
        reprovados += 1

# mostra o total
print(f"\n3Aprovados: {aprovados}")
print(f"Recuperação: {recuperacao}")
print(f"Reprovados: {reprovados}")
```

```python
# 27) Par ou Ímpar até N

# para ler o o valor de N
n = int(input("Digite um número: "))

# vê se o número é válido
if n > 0:

    # vai de 1 até N
    for i in range(1, n + 1):

        # vê se o número é par
        if i % 2 == 0:
            print(f"{i} é par")

        # se não for par, é ímpar
        else:
            print(f"{i} é ímpar")

else:
    print("Número inválido")
```

```python
# 28) Caça aos Primos

# pede o primeiro número
numero = int(input("Digite um número (-1 para encerrar): "))

# variáveis
quantidade_primos = 0
soma_nao_primos = 0

# enquanto não digitar -1
while numero != -1:
    divisores = 0

    # para ver quantos divisores o número possui
    for i in range(1, numero + 1):

        if numero % i == 0:
            divisores += 1

    # número primo tem apenas 2 divisores
    if divisores == 2:
        quantidade_primos += 1

    else:
        soma_nao_primos += numero

    # pede o próximo número
    numero = int(input("Digite um número (-1 para encerrar): "))

# mostra os resultados
print(f"\nQuantidade de números primos: {quantidade_primos}")
print(f"Soma dos números não primos: {soma_nao_primos}")
```

```python
# 29) Avaliação de Notas

# pede a primeira nota
nota = float(input("Digite uma nota (0 para encerrar): "))

# variáveis
soma = 0
contador = 0

# enquanto não digitar 0
while nota != 0:
    soma += nota
    contador += 1

    # pede a próxima nota
    nota = float(input("Digite uma nota (0 para encerrar): "))

# para calcular a média
if contador > 0:
    media = soma / contador
    acima_media = 0

    print(f"\nMédia geral: {media:.2f}")

    # pede as notas novamente
    print("\nDigite as notas novamente para verificar quais estão acima da média: ")

    for i in range(1, contador + 1):
        n = float(input(f"Digite a {i}° nota: "))

        # para ver se está acima da média
        if n > media:
            acima_media += 1

    # mostra o resultado
    print(f"\nNotas acima da média: {acima_media}")

else:
    print("Nenhuma nota foi digitada")
```

```python
# 30) Cálculo de Fatorial

# ler o número
n = int(input("Digite um número inteiro positivo: "))

# para ver se o número é válido
if n >= 0:

    # começa o fatorial em 1
    fatorial = 1

    # calcula o fatorial
    for i in range(1, n + 1):
        fatorial *= i

    # mostra o resultado do fatorial
    print(f"Fatorial de {n}: {fatorial}")

    # verifica se o fatorial é par ou ímpar
    if fatorial % 2 == 0:
        print("O fatorial é par")

    else:
        print("O fatorial é ímpar")

else:
    print("Número inválido")
```
