```python
# ATIVIDADE 7

# Primeira lista
lista1 = []

# Segunda lista
lista2 = []

# Leitura da primeira lista
for i in range(10):
    numero = int(input(f'Digite o {i+1}° número da lista 1: '))
    lista1.append(numero)

# Leitura da segunda lista
for i in range(10):
    numero = int(input(f'Digite o {i+1}° número da lista 2: '))
    lista2.append(numero)

# Lista da interseção
intersecao = []

# Verifica quais números estão presentes nas duas listas
for numero in lista1:

    if numero in lista2:

        # Evita números repetidos na interseção
        if numero not in intersecao:
            intersecao.append(numero)

# Exibe os resultados
print('Lista 1:', lista1)
print('Lista 2:', lista2)
print('Interseção:', intersecao)
```

```python
# ATIVIDADE 16

# Criando a matriz vazia
matriz = []

# Lendo os valores da matriz
for i in range(4):
    linha = []

    for j in range(4):
        numero = int(input(f'Digite o valor [{i}][{j}]: '))
        linha.append(numero)

    matriz.append(linha)

# Lista que armazenará os menores valores
menores = []

# Percorrendo cada coluna
for coluna in range(4):

    # Assume o primeiro valor da coluna como menor
    menor = matriz[0][coluna]

    # Verifica os outros valores da coluna
    for linha in range(4):

        if matriz[linha][coluna] < menor:
            menor = matriz[linha][coluna]

    # Adiciona o menor valor encontrado
    menores.append(menor)

print("Menores valores de cada coluna:")
print(menores)
```

```python
# ATIVIDADE 17

# Criando a matriz
matriz = []

# Lendo os valores da matriz
for i in range(3):

    linha = []

    for j in range(3):

        valor = float(input(f'Digite o valor [{i}][{j}]: '))
        linha.append(valor)

    matriz.append(linha)

# Diagonais principais
d1 = matriz[0][0] * matriz[1][1] * matriz[2][2]
d2 = matriz[0][1] * matriz[1][2] * matriz[2][0]
d3 = matriz[0][2] * matriz[1][0] * matriz[2][1]

# Diagonais secundárias
d4 = matriz[0][2] * matriz[1][1] * matriz[2][0]
d5 = matriz[0][0] * matriz[1][2] * matriz[2][1]
d6 = matriz[0][1] * matriz[1][0] * matriz[2][2]

# Calculando o determinante
determinante = (d1 + d2 + d3) - (d4 + d5 + d6)

# Exibindo o resultado
print("Determinante =", determinante)
```

```python
# ATIVIDADE 18

# Primeira matriz
matriz1 = []

for i in range(3):
    linha = []

    for j in range(3):
        valor = int(input(f'Matriz 1 [{i}][{j}]: '))
        linha.append(valor)

    matriz1.append(linha)

# Segunda matriz
matriz2 = []

for i in range(3):
    linha = []

    for j in range(3):
        valor = int(input(f'Matriz 2 [{i}][{j}]: '))
        linha.append(valor)

    matriz2.append(linha)

# Matrizes resultado
soma = []
diferenca = []

# Calculando soma e diferença
for i in range(3):

    linha_soma = []
    linha_diferenca = []

    for j in range(3):

        linha_soma.append(matriz1[i][j] + matriz2[i][j])

        linha_diferenca.append(matriz1[i][j] - matriz2[i][j])

    soma.append(linha_soma)
    diferenca.append(linha_diferenca)

print("Matriz Soma:")
for linha in soma:
    print(linha)

print("Matriz Diferença:")
for linha in diferenca:
    print(linha)
```

```python
# ATIVIDADE 19

# Listas para receber os números digitados
listaA = []
listaB = []

# Lendo os elementos de A
for i in range(10):

    numero = int(input(f'Digite o {i+1}° elemento de A: '))
    listaA.append(numero)

# Lendo os elementos de B
for i in range(10):

    numero = int(input(f'Digite o {i+1}° elemento de B: '))
    listaB.append(numero)

# Convertendo as listas para conjuntos
A = set(listaA)
B = set(listaB)

# Operações entre conjuntos
intersecao = A & B
uniao = A | B
diferencaAB = A - B
diferencaBA = B - A

# Exibindo os resultados
print("Interseção:", intersecao)
print("União:", uniao)
print("A - B:", diferencaAB)
print("B - A:", diferencaBA)
```

```python
# ATIVIDADE 23

# Criando os conjuntos
A = set()
B = set()

# Lendo os elementos de A
quantidadeA = int(input("Quantos elementos terá o conjunto A? "))

for i in range(quantidadeA):

    numero = int(input(f'Elemento {i+1} de A: '))
    A.add(numero)

# Lendo os elementos de B
quantidadeB = int(input("Quantos elementos terá o conjunto B? "))

for i in range(quantidadeB):

    numero = int(input(f'Elemento {i+1} de B: '))
    B.add(numero)

# Verificando se A é subconjunto de B
subconjuntoA = True

for elemento in A:

    if elemento not in B:
        subconjuntoA = False

# Verificando se B é subconjunto de A
subconjuntoB = True

for elemento in B:

    if elemento not in A:
        subconjuntoB = False

# Exibindo o resultado
if subconjuntoA:

    print("A é subconjunto de B")

elif subconjuntoB:

    print("B é subconjunto de A")

else:

    print("Nenhum conjunto é subconjunto do outro")
```

```python
# ATIVIDADE 24

# Lista para armazenar os números
numeros = []

# Lendo os números
quantidade = int(input("Quantos números deseja informar? "))

for i in range(quantidade):

    numero = int(input(f'Digite o {i+1}° número: '))
    numeros.append(numero)

# Lista para os quadrados perfeitos
quadrados_perfeitos = []

# Verificando cada número
for numero in numeros:

    raiz = int(numero ** 0.5)

    # Se raiz² for igual ao número, ele é quadrado perfeito
    if raiz * raiz == numero:

        # Evita repetições
        if numero not in quadrados_perfeitos:

            quadrados_perfeitos.append(numero)

# Transformando em conjunto
quadrados_perfeitos = set(quadrados_perfeitos)

# Exibindo o resultado
print("Quadrados perfeitos encontrados:")
print(quadrados_perfeitos)
```

```python
4# ATIVIDADE 25

# Criando o dicionário
funcionarios = {}

# Lendo os dados dos funcionários
for i in range(5):

    nome = input(f'Digite o nome do {i+1}° funcionário: ')
    salario = float(input("Digite o salário: R$ "))

    funcionarios[nome] = salario

# Aplicando o reajuste
for nome, salario in funcionarios.items():

    # Verificando se o salário é menor que 2000
    if salario < 2000:

        # Aumentando 10%
        novo_salario = salario * 1.10

        # Atualizando o dicionário
        funcionarios[nome] = novo_salario

# Exibindo os salários atualizados
print("\nSalários atualizados:")

for nome, salario in funcionarios.items():

    print(f'{nome}: R$ {salario:.2f}')
```

```python
# ATIVIDADE 26

# Criando o dicionário
alunos = {}

# Lendo os dados dos alunos
for i in range(5):

    nome = input(f'Digite o nome do {i+1}° aluno: ')

    notas = []

    # Lendo as 3 notas
    for j in range(3):

        nota = float(input(f'Digite a {j+1}ª nota: '))
        notas.append(nota)

    # Armazenando no dicionário
    alunos[nome] = notas

# Exibindo as médias
for nome, notas in alunos.items():

    soma = 0

    # Somando as notas
    for nota in notas:
        soma += nota

    media = soma / len(notas)

    print(f'Aluno: {nome}')
    print(f'Média: {media:.2f}\n')
```

```python
7# ATIVIDADE 29

# Criando o dicionário
produtos = {}

# Lendo os produtos
for i in range(3):

    nome = input(f'Digite o nome do {i+1}° produto: ')

    precos = []

    # Lendo os preços das 3 lojas
    for j in range(3):

        preco = float(input(f'Preço na loja {j+1}: R$ '))
        precos.append(preco)

    produtos[nome] = precos

# Variáveis de controle
maior_variacao = 0
produto_maior = ""

# Verificando cada produto
for produto, precos in produtos.items():

    maior = precos[0]
    menor = precos[0]

    # Encontrando maior e menor preço
    for preco in precos:

        if preco > maior:
            maior = preco

        if preco < menor:
            menor = preco

    variacao = maior - menor

    # Atualizando a maior variação encontrada
    if variacao > maior_variacao:

        maior_variacao = variacao
        produto_maior = produto

print(f'Produto com maior variação: {produto_maior}')
print(f'Variação: R$ {maior_variacao:.2f}')
```

```python
# ATIVIDADE 30

# Lendo a frase
frase = input("Digite uma frase: ")

# Convertendo para minúsculas
frase = frase.lower()

# Separando as palavras
palavras = frase.split()

# Criando o dicionário
ocorrencias = {}

# Contando as ocorrências
for palavra in palavras:

    # Se a palavra ainda não existe
    if palavra not in ocorrencias:

        ocorrencias[palavra] = 1

    else:

        ocorrencias[palavra] += 1

# Exibindo o resultado
for palavra, quantidade in ocorrencias.items():

    print(f'{palavra}: {quantidade}')
```
