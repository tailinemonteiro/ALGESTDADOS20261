```python
# 1) Classificador de Três Números

# lê os números, n1= "número 1" e assim sucessivamente, para facilitar:)
n1 = int(input("Número 1: "))
n2 = int(input("Número 2: "))
n3 = int(input("Número 3: "))

# maior
if n1 >= n2 and n1 >= n3:
    maior = n1
elif n2 >= n1 and n2 >= n3:
    maior = n2
else:
    maior = n3

# menor
if n1 <= n2 and n1 <= n3:
    menor = n1
elif n2 <= n1 and n2 <= n3:
    menor = n2
else:
    menor = n3

# do meio
meio = n1 + n2 + n3 - maior - menor

# mostra resultado
print("Maior:", maior)
print("Meio:", meio)
print("Menor:", menor)

# verifica se tem números iguais
if n1 == n2 or n1 == n3 or n2 == n3:
    print("Obs: existem números iguais")
```

```python
# 2) Calculadora de Gorjeta

# pede os dados
conta = float(input("Digite o valor da conta: "))
nota = int(input("Nota do serviço (1 a 5): "))

# validação
if conta <= 0 or nota < 1 or nota > 5:
    print("Dados inválidos!")
else:

    # define a porcentagem da gorjeta
    if nota == 5:
        porcentagem = 0.20
    elif nota == 4:
        porcentagem = 0.15
    elif nota == 3:
        porcentagem = 0.10
    else:  # nota 1 ou 2
        porcentagem = 0

    # calcula a gorjeta
    gorjeta = conta * porcentagem

    # total a pagar
    total = conta + gorjeta

    # mostra os resultados
    print(f"Gorjeta: R$ {gorjeta:.2f}")
    print(f"Total a pagar: R$ {total:.2f}")
```

```python
# 3) Verificador de Ano Bissexto

# lê o ano
ano = int(input("Digite o ano: "))

# validação
if ano <= 0:
    print("Ano inválido!")
else:

    # começa assumindo que não é bissexto
    bissexto = False

    # regras do ano bissexto (na ordem correta)
    if (ano % 400 == 0):
        bissexto = True
    else:
        if (ano % 100 == 0):
            bissexto = False
        else:
            if (ano % 4 == 0):
                bissexto = True
            else:
                bissexto = False

    # resultado
    if bissexto:
        print("Ano bissexto")
    else:
        print("Ano não bissexto")
```

```python
# 4) Cálculo de Imposto de Renda Simplificado

# lê o salário
salario_bruto = float(input("Digite o salário bruto: "))

# validação
if salario_bruto <= 0:
    print("Salário inválido!")
else:

    # define a alíquota
    if salario_bruto <= 2259:
        aliquota = 0
    elif salario_bruto <= 2826:
        aliquota = 0.075
    elif salario_bruto <= 3751:
        aliquota = 0.15
    elif salario_bruto <= 4664:
        aliquota = 0.225
    else:
        aliquota = 0.275

    # calcula o imposto
    imposto = salario_bruto * aliquota

    # salário líquido
    salario_liquido = salario_bruto - imposto

    # mostra os resultados
    print(f"Alíquota aplicada: {aliquota * 100:.2f}%")
    print(f"Imposto: R$ {imposto:.2f}")
    print(f"Salário líquido: R$ {salario_liquido:.2f}")
```

```python
# 5) Placar de Futebol

# lê os gols
gols_a = int(input("Gols do time A: "))
gols_b = int(input("Gols do time B: "))

# validação
if gols_a < 0 or gols_b < 0:
    print("Valores inválidos!")
else:

    # calcula a diferença de gols
    diferenca = gols_a - gols_b

    # verifica quem venceu
    if gols_a > gols_b:
        print("Time A venceu")
        diferenca = gols_a - gols_b

        # verifica goleada
        if diferenca > 3:
            print("Foi goleada!")

    elif gols_b > gols_a:
        print("Time B venceu")
        diferenca = gols_b - gols_a

        # verifica goleada
        if diferenca > 3:
            print("Foi goleada!")

    else:
        print("Empate")
        diferenca = 0

    # mostra a diferença
    print("Diferença de gols:", diferenca)
```

```python
# 6) Validador de Data

# lê os dados
dia = int(input("Digite o dia: "))
mes = int(input("Digite o mês: "))
ano = int(input("Digite o ano: "))

# valida ano e mês
if ano <= 0:
    print("Ano inválido!")
elif mes < 1 or mes > 12:
    print("Mês inválido!")
else:

    # verifica se é ano bissexto
    if (ano % 400 == 0):
        bissexto = True
    else:
        if (ano % 100 == 0):
            bissexto = False
        else:
            if (ano % 4 == 0):
                bissexto = True
            else:
                bissexto = False

    # agora verifica o dia dependendo do mês
    if mes == 4 or mes == 6 or mes == 9 or mes == 11:
        # meses com 30 dias
        if dia < 1 or dia > 30:
            print("Dia inválido para esse mês!")
        else:
            print("Data válida!")

    elif mes == 2:
        # fevereiro
        if bissexto:
            if dia < 1 or dia > 29:
                print("Dia inválido para fevereiro (bissexto)!")
            else:
                print("Data válida!")
        else:
            if dia < 1 or dia > 28:
                print("Dia inválido para fevereiro!")
            else:
                print("Data válida!")

    else:
        # meses com 31 dias
        if dia < 1 or dia > 31:
            print("Dia inválido para esse mês!")
        else:
            print("Data válida!")
```

```python
# 7) Calculadora de Frete

# pede os dados
peso = float(input("Digite o peso (kg): "))
regiao = int(input("Código da região (1 = Sul/Sudeste, 2 = Centro-Oeste/Nordeste, 3 = Norte): "))

# validação
if peso <= 0 or regiao < 1 or regiao > 3:
    print("Dados inválidos!")
else:

    # define o frete com base no peso
    if peso <= 1:
        frete = 15
    elif peso <= 5:
        frete = 25
    else:
        frete = 40

    # aplica acréscimo do valor para as regiões 2 e 3
    if regiao == 2 or regiao == 3:
        frete = frete + (frete * 0.20)

    # resultado
    print(f"Frete final: R$ {frete:.2f}")
```

```python
# 8) Simulador de Caixa Eletrônico

# dados fixos do sistema
saldo = 1000
senha_correta = "1234"

# entradas
valor = float(input("Valor para saque: "))
senha = input("Digite a senha: ")

# validação do valor
if valor <= 0:
    print("Valor inválido!")
else:

    # verifica a senha
    if senha != senha_correta:
        print("Senha incorreta!")
    else:

        # verifica saldo
        if valor > saldo:
            print("Saldo insuficiente!")
        else:

            # saque aprovado
            saldo = saldo - valor
            print("Saque realizado com sucesso!")
            print(f"Saldo restante: R$ {saldo:.2f}")
```

```python
# 9) Conversor de Temperatura Completo

# lê os dados
temperatura = float(input("Digite a temperatura: "))
unidade = input("Unidade (C, F ou K): ").upper()

# usa match/case para identificar a unidade
match unidade:
    case "C":
        # já está em Celsius
        temp_celsius = temperatura

        # converte para Fahrenheit
        temp_fahrenheit = (temp_celsius * 9/5) + 32

        # converte para Kelvin
        temp_kelvin = temp_celsius + 273

        print(f"Celsius: {temp_celsius:.2f}")
        print(f"Fahrenheit: {temp_fahrenheit:.2f}")
        print(f"Kelvin: {temp_kelvin:.2f}")

    case "F":
        # converte para Celsius
        temp_celsius = (temperatura - 32) * 5/9

        # converte para Kelvin
        temp_kelvin = temp_celsius + 273

        temp_fahrenheit = temperatura

        print(f"Celsius: {temp_celsius:.2f}")
        print(f"Fahrenheit: {temp_fahrenheit:.2f}")
        print(f"Kelvin: {temp_kelvin:.2f}")

    case "K":
        # validação de Kelvin
        if temperatura < 0:
            print("Temperatura em Kelvin não pode ser negativa!")
        else:
            temp_kelvin = temperatura

            # converte para Celsius
            temp_celsius = temp_kelvin - 273

            # converte para Fahrenheit
            temp_fahrenheit = (temp_celsius * 9/5) + 32

            print(f"Celsius: {temp_celsius:.2f}")
            print(f"Fahrenheit: {temp_fahrenheit:.2f}")
            print(f"Kelvin: {temp_kelvin:.2f}")

    case _:
        print("Unidade inválida!")
```

```python
# 10) Classificador de IMC com Risco

# lê os dados
peso = float(input("Digite o peso (kg): "))
altura = float(input("Digite a altura (m): "))

# validação
if peso <= 0 or altura < 0.5 or altura > 2.5:
    print("Valores inválidos!")
else:

    # calcula o IMC
    imc = peso / (altura * altura)

    # classificação do IMC
    if imc < 18.5:
        classificacao = "Abaixo do peso"

        # riscos atrelados acada classificação
        risco = "Baixo"

    elif imc < 25:
        classificacao = "Normal"

        risco = "Baixo"

    elif imc < 30:
        classificacao = "Sobrepeso"

        risco = "Moderado"

    elif imc < 35:
        classificacao = "Obesidade Grau I"

        risco = "Moderado"

    elif imc < 40:
        classificacao = "Obesidade Grau II"

        risco = "Alto"

    else:
        classificacao = "Obesidade Grau III"

        risco = "Alto"

    # resultado
    print(f"IMC: {imc:.2f}")
    print("Classificação:", classificacao)
    print("Risco cardiovascular:", risco)
```

```python
# 11) Detector de Múltiplos

# lê o número
numero = int(input("Digite um número inteiro positivo: "))

# validação
if numero <= 0:
    print("Número inválido!")
else:

    # verifica os múltiplos
    e_multiplo_de_2 = (numero % 2 == 0)
    e_multiplo_de_3 = (numero % 3 == 0)
    e_multiplo_de_5 = (numero % 5 == 0)

    # mostra quais são verdadeiros
    if e_multiplo_de_2:
        print("É múltiplo de 2")

    if e_multiplo_de_3:
        print("É múltiplo de 3")

    if e_multiplo_de_5:
        print("É múltiplo de 5")

    # verifica se é múltiplo dos três ao mesmo tempo
    if (e_multiplo_de_2 and e_multiplo_de_3 and e_multiplo_de_5):
        print("É múltiplo de 2, 3 e 5 ao mesmo tempo")
```

```python
# 12) Calculadora de Desconto Progressivo

# lê os dados
preco = float(input("Digite o preço do produto: "))
cupom = input("Digite o cupom (PROMO10, PROMO20, PROMO30, SEMCUPOM): ").upper()

# começa sem desconto
percentual = 0

# define o desconto pelo cupom
match cupom:
    case "PROMO10":
        percentual = 0.10
    case "PROMO20":
        percentual = 0.20
    case "PROMO30":
        percentual = 0.30
    case "SEMCUPOM":
        percentual = 0
    case _:
        print("Cupom inválido!")
        percentual = 0

# verifica limite de preço baixo
if preco < 50:

    # limita o desconto a no máximo 10%
    if percentual > 0.10:
        percentual = 0.10

# calcula o desconto em reais
desconto = preco * percentual

# preço final
preco_final = preco - desconto

# mostra resultado
print("Cupom usado:", cupom)
print(f"Desconto: R$ {desconto:.2f}")
print(f"Preço final: R$ {preco_final:.2f}")
```

```python
# 13) Avaliador de Triângulo Completo

# lê os lados, l1= "lado 1" e assim sucessivamente, para facilitar:)
l1 = float(input("Lado 1: "))
l2 = float(input("Lado 2: "))
l3 = float(input("Lado 3: "))

# validação somente com valores positivos
if l1 <= 0 or l2 <= 0 or l3 <= 0:
    print("Valores inválidos!")
else:

    # verifica a desigualdade triangular
    if (l1 < l2 + l3) and (l2 < l1 + l3) and (l3 < l1 + l2):

        # classificação
        if l1 == l2 and l2 == l3:
            tipo = "Equilátero"
        elif l1 == l2 or l1 == l3 or l2 == l3:
            tipo = "Isósceles"
        else:
            tipo = "Escaleno"

        print("Triângulo válido")
        print("Tipo:", tipo)

    else:
        print("Triângulo inválido (não atende à desigualdade triangular)")
```

```python
# 14) Calculadora de Notas com Conceito

# lê as notas, n1= "nota 1" e assim sucessivamente, para facilitar:)
n1 = float(input("Nota 1: "))
n2 = float(input("Nota 2: "))
n3 = float(input("Nota 3: "))
n4 = float(input("Nota 4: "))

# validação das notas, todas entre 0 e 108
if (n1 < 0 or n1 > 10) or (n2 < 0 or n2 > 10) or (n3 < 0 or n3 > 10) or (n4 < 0 or n4 > 10):
    print("Notas inválidas!")
else:

    # calcula a média
    media = (n1 + n2 + n3 + n4) / 4

    # define o conceito
    if media >= 9:
        conceito = "A"
    elif media >= 7:
        conceito = "B"
    elif media >= 5:
        conceito = "C"
    elif media >= 3:
        conceito = "D"
    else:
        conceito = "E"

    # define o status
    if conceito == "A" or conceito == "B":
        status = "Aprovado"
    elif conceito == "C":
        status = "Recuperação"
    else:
        status = "Reprovado"

    # resultado
    print(f"Média: {media:.2f}")
    print("Conceito:", conceito)
    print("Status:", status)
```

```python
# 15) Verificador de Acesso por Horário

# lê os dados
hora = int(input("Digite a hora (0 a 23): "))
tipo = input("Tipo de usuário (admin, operador, visitante): ").lower()

# valida a hora
if hora < 0 or hora > 23:
    print("Hora inválida!")
else:

    # define o turno
    if hora >= 6 and hora <= 11:
        turno = "Manhã"
    elif hora <= 17:
        turno = "Tarde"
    elif hora <= 23:
        turno = "Noite"
    else:
        turno = "Madrugada"

    # verifica acesso pelo tipo de cargo
    match tipo:
        case "admin":
            acesso = "Permitido"

        case "operador":
            # operador só manhã e tarde
            if turno == "Manhã" or turno == "Tarde":
                acesso = "Permitido"
            else:
                acesso = "Negado"

        case "visitante":
            # visitante só tarde
            if turno == "Tarde":
                acesso = "Permitido"
            else:
                acesso = "Negado"

        case _:
            print("Tipo de usuário inválido!")
            acesso = ""

    # mostra resultado
    if acesso != "":
        print("Turno:", turno)
        print("Acesso:", acesso)
```

```python
# 16) Simulador de Bomba de Combustível

# preços de cada combustível
preco_gasolina = 5.80
preco_etanol = 4.20
preco_diesel = 6.10

# pede o tipo de combustível
tipo = input("Digite o combustível (gasolina, etanol ou diesel): ")

# pede a quantidade de litros
litros = float(input("Quantos litros você quer abastecer? "))

# primeiro: ver se a quantidade faz sentido
if litros <= 0:
    print("Quantidade inválida, precisa ser maior que zero.")
else:

    # aqui ele identifica qual combustível foi escolhido
    match tipo:
        case "gasolina":
            preco = preco_gasolina
        case "etanol":
            preco = preco_etanol
        case "diesel":
            preco = preco_diesel
        case _:
            print("Combustível inválido!")
            preco = 0  # coloquei 0 só pra não dar erro depois

    # só continua se o tipo estiver certo
    if preco != 0:

        # calcula o valor sem desconto
        subtotal = litros * preco

        # começa sem desconto
        desconto = 0

        # se abastecer mais de 40L, ganha 3% de desconto
        if litros > 40:
            desconto = subtotal * 0.03

        # valor final
        total = subtotal - desconto

        # mostra os resultados
        print(f"Subtotal: R$ {subtotal:.2f}")
        print(f"Desconto: R$ {desconto:.2f}")
        print(f"Total a pagar: R$ {total:.2f}")
```

```python
# 17) Classificador de Velocidade

# pede a velocidade
velocidade = float(input("Digite a velocidade (km/h): "))

# pede o tipo de via
via = input("Digite o tipo de via (urbana, rodovia, autopista): ").lower()

# primeiro: validar a velocidade
if velocidade < 0:
    print("Velocidade inválida!")
else:

    # define o limite de acordo com a via
    match via:
        case "urbana":
            limite = 60
        case "rodovia":
            limite = 100
        case "autopista":
            limite = 120
        case _:
            print("Tipo de via inválido!")
            limite = 0

    # só continua se a via for válida
    if limite != 0:

        # começa assumindo que está dentro do limite
        situacao = ""

        # agora compara a velocidade com o limite
        if velocidade <= limite:
            situacao = "Dentro do limite"

        # até 20% acima do limite
        elif velocidade <= limite * 1.2:
            situacao = "Excesso leve"

        # até 50% acima do limite
        elif velocidade <= limite * 1.5:
            situacao = "Excesso moderado"

        # acima de 50%
        else:
            situacao = "Excesso grave"

        # mostra os resultados
        print(f"Limite da via: {limite} km/h")
        print(f"Sua velocidade: {velocidade} km/h")
        print(f"Situação: {situacao}")
```

```python
# 18) Calculadora de Parcelas com Juros

# pede os dados
valor = float(input("Digite o valor da compra: "))
parcelas = int(input("Digite o número de parcelas (1 a 12): "))

# validação
if valor <= 0:
    print("Valor inválido!")
elif parcelas < 1 or parcelas > 12:
    print("Parcelas inválidas!")
else:

    # define os juros
    if parcelas == 1:
        juros = 0
    elif parcelas <= 3:
        juros = 0.05
    elif parcelas <= 6:
        juros = 0.10
    else:
        juros = 0.15

    # calcula total com juros
    total = valor + (valor * juros)

    # valor de cada parcela
    parcela = total / parcelas

    # mostra tudo
    print("Total com juros:", total)
    print("Valor da parcela:", parcela)

    # verifica se a parcela é muito pequena
    if parcela < 20:
        print("Parcela menor que R$20!")

        parcelas_min = int(total / 20) + 1
        print("Número mínimo de parcelas:", parcelas_min)
```

```python
# 19) Detector de Intervalo com Tolerância

# pede os valores
alvo = float(input("Digite o valor alvo: "))
tolerancia = float(input("Digite a tolerância: "))
medido = float(input("Digite o valor medido: "))

# validação (todos precisam ser positivos)
if alvo <= 0 or tolerancia < 0 or medido <= 0:
    print("Valores inválidos!")
else:

    # calcula os limites
    limite_inferior = alvo - tolerancia
    limite_superior = alvo + tolerancia

    # calcula o desvio (diferença do medido pro alvo)
    desvio = medido - alvo

    # verifica onde o valor medido está
    if medido < limite_inferior:
        situacao = "Abaixo do limite"
    elif medido > limite_superior:
        situacao = "Acima do limite"
    else:
        situacao = "Dentro da faixa"

    # mostra os resultados
    print("Limite inferior:", limite_inferior)
    print("Limite superior:", limite_superior)
    print("Valor medido:", medido)
    print("Desvio:", desvio)
    print("Classificação:", situacao)
```

```python
# 20) Simulador de Pedágio Inteligente

# tarifas base de cada veículo
moto = 5
carro = 10
caminhao = 20
onibus = 15

# tipo de veículo
tipo = input("Digite o tipo de veículo (moto, carro, caminhao, onibus): ")

# se tem tag
tag = input("Possui tag? (s/n): ").lower()

# hora atual
hora = int(input("Digite a hora (0 a 23): "))

# validação da hora
if hora < 0 or hora > 23:
    print("Hora inválida!")
else:

    # define a tarifa base
    match tipo:
        case "moto":
            tarifa = moto
        case "carro":
            tarifa = carro
        case "caminhao":
            tarifa = caminhao
        case "onibus":
            tarifa = onibus
        case _:
            print("Tipo de veículo inválido!")
            tarifa = 0  # pra não dar erro

    # só continua se o tipo for válido
    if tarifa != 0:

        valor = tarifa

        # desconto se tiver tag
        if tag == "s":
            valor = valor - (valor * 0.10)  # tira 10%

        # verifica horário de pico
        if (hora >= 7 and hora <= 9) or (hora >= 17 and hora <= 19):
            valor = valor + (valor * 0.20)  # acrescenta 20%

        # mostra os resultados
        print("Tarifa base:", tarifa)
        print("Valor final:", valor)
```

```python
# 21) Calculadora de Energia Elétrica

# pede o consumo
consumo = float(input("Digite o consumo em kWh: "))

# validação
if consumo < 0:
    print("Consumo inválido!")
else:

    # define o preço por kWh
    if consumo <= 100:
        preco = 0.50
    elif consumo <= 200:
        preco = 0.75
    elif consumo <= 300:
        preco = 1.00
    else:
        preco = 1.50

    # calcula o valor sem desconto
    total = consumo * preco

    # começa sem desconto
    desconto = 0

    # só pergunta do modo econômico se passar de 300
    if consumo > 300:
        modo = input("Modo econômico? (s/n): ").lower()

        # se respondeu sim, aplica 10% de desconto
        if modo == "s":
            desconto = total * 0.10

    # valor final
    total_final = total - desconto

    # mostra os resultados
    print("Valor sem desconto:", total)
    print("Desconto:", desconto)
    print("Valor final:", total_final)
```

```python
# 22) Verificador de Intervalo Múltiplo

# pede os valores
A = float(input("Digite o valor mínimo (A): "))
B = float(input("Digite o valor máximo (B): "))
N = float(input("Digite o número (N): "))
D = float(input("Digite o divisor (D): "))

# validações
if A >= B:
    print("Intervalo inválido! A deve ser menor que B.")
elif D <= 0:
    print("Divisor inválido!")
else:

    # verifica se N está dentro do intervalo
    if N >= A and N <= B:

        # agora verifica se é divisível
        if N % D == 0:
            print("Dentro do intervalo e divisível")
        else:
            print("Dentro do intervalo e não divisível")

    else:

        # ainda precisa verificar se é divisível
        if N % D == 0:
            print("Fora do intervalo e divisível")
        else:
            print("Fora do intervalo e não divisível")
```

```python
# 23) Sistema de Pontuação de Jogo

# entradas, j1= "jogador 1" e assim sucessivamente, para facilitar:)
j1 = int(input("Jogador 1: "))
j2 = int(input("Jogador 2: "))
j3 = int(input("Jogador 3: "))

# validação
if j1 < 0 or j1 > 1000 or j2 < 0 or j2 > 1000 or j3 < 0 or j3 > 1000:
    print("Valores inválidos!")
else:

    # maior pontuação
    if j1 >= j2 and j1 >= j3:
        maior = j1
    elif j2 >= j1 and j2 >= j3:
        maior = j2
    else:
        maior = j3

    # menor pontuação
    if j1 <= j2 and j1 <= j3:
        menor = j1
    elif j2 <= j1 and j2 <= j3:
        menor = j2
    else:
        menor = j3

    # o do meio
    meio = j1 + j2 + j3 - maior - menor

    # diferença
    diferenca = maior - menor

    # resultado
    print("1º lugar:", maior)
    print("2º lugar:", meio)
    print("3º lugar:", menor)
    print("Diferença:", diferenca)

    # só um aviso simples de empate
    if j1 == j2 or j1 == j3 or j2 == j3:
        print("Obs: houve empate entre jogadores")
```

```python
# 24) Calculadora de Reajuste Salarial

# pede os dados
salario = float(input("Digite o salário atual: "))
avaliacao = float(input("Digite a avaliação (0 a 100): "))

# validação
if salario <= 0 or avaliacao < 0 or avaliacao > 100:
    print("Dados inválidos!")
else:

    # define o percentual de reajuste
    if avaliacao >= 90:
        percentual = 0.15
    elif avaliacao >= 75:
        percentual = 0.10
    elif avaliacao >= 60:
        percentual = 0.05
    else:
        percentual = 0

    # aumento em reais
    aumento = salario * percentual

    # novo salário
    novo_salario = salario + aumento

    # verifica o teto
    if novo_salario > 15000:
        novo_salario = 15000
        print("Salário ajustado ao teto de R$15.000")

    # mostra os resultados
    print("Salário antigo:", salario)
    print("Percentual aplicado:", percentual * 100, "%")
    print("Aumento:", aumento)
    print("Novo salário:", novo_salario)
```

```python
# 25) Classificador de Triângulo com Ângulos

# pede os ângulos, a1= "ângulo 1" e assim sucessivamente, para facilitar:)
a1 = float(input("Ângulo 1: "))
a2 = float(input("Ângulo 2: "))
a3 = float(input("Ângulo 3: "))

# validação
if a1 <= 0 or a2 <= 0 or a3 <= 0:
    print("Ângulos inválidos!")
else:

    # soma dos ângulos
    soma = a1 + a2 + a3

    # verifica se forma triângulo
    if soma != 180:
        print("Triângulo inválido")
        print("Soma encontrada:", soma)
    else:

        # classificação pelos ângulos
        if a1 == 90 or a2 == 90 or a3 == 90:
            tipo_angulo = "Retângulo"
        elif a1 > 90 or a2 > 90 or a3 > 90:
            tipo_angulo = "Obtusângulo"
        else:
            tipo_angulo = "Acutângulo"

        # classificação pelos lados (usando os ângulos iguais)
        if a1 == a2 and a2 == a3:
            tipo_lado = "Equilátero"
        elif a1 == a2 or a1 == a3 or a2 == a3:
            tipo_lado = "Isósceles"
        else:
            tipo_lado = "Escaleno"

        # mostra resultado
        print("Tipo quanto aos ângulos:", tipo_angulo)
        print("Tipo quanto aos lados:", tipo_lado)
```

```python
# 26) Calculadora de Combustível de Viagem

# pede os dados
distancia = float(input("Digite a distância (km): "))
consumo = float(input("Consumo do veículo (km/l): "))
preco = float(input("Preço do combustível (R$/l): "))

# validação
if distancia <= 0 or consumo <= 0 or preco <= 0:
    print("Valores inválidos!")
else:

    # calcula quantos litros vai gastar
    litros = distancia / consumo

    # calcula o custo total
    custo = litros * preco

    # mostra os dados principais
    print(f"Distância: {distancia:.2f} km")
    print(f"Litros necessários: {litros:.2f} L")
    print(f"Custo total: R$ {custo:.2f}")

    # verifica se passou de 500
    if custo > 500:
        print("Custo alto! Recomenda-se uma parada.")

        # calcula o ponto no meio da viagem
        parada = distancia / 2

        print(f"Parada sugerida no km: {parada:.2f}")
```

```python
# 27) Verificador de Senha com Nível de Segurança

# pede a senha e a confirmação
senha = input("Digite a senha: ")
confirmacao = input("Confirme a senha: ")

# verifica se são iguais
if senha != confirmacao:
    print("Senhas não conferem")
else:

    # define o nível pelo tamanho
    if len(senha) < 6:
        nivel = "Fraca"
    elif len(senha) <= 9:
        nivel = "Média"
    else:
        nivel = "Forte"

    # verifica se tem só números
    so_numeros = True

    for c in senha:
        if c < "0" or c > "9":  # se achar algo que não é número
            so_numeros = False

    # se for só número, rebaixa o nível
    if so_numeros:
        if nivel == "Forte":
            nivel = "Média"
        elif nivel == "Média":
            nivel = "Fraca"

    # resultado final
    print("Nível da senha:", nivel)
```

```python
# 28) Simulador de Reserva de Assento

# pede os dados
assento = int(input("Digite o número do assento (1 a 100): "))
classe = input("Digite a classe (economica, executiva, primeira): ").lower()

# valida o número do assento
if assento < 1 or assento > 100:
    print("Assento inválido!")
else:

    # define o preço pela classe
    match classe:
        case "economica":
            preco = 350
        case "executiva":
            preco = 890
        case "primeira":
            preco = 1500
        case _:
            print("Classe inválida!")
            preco = 0  # evita erro

    # só continua se a classe for válida
    if preco != 0:

        # verifica se o assento bate com a classe
        if assento >= 1 and assento <= 10:
            faixa = "primeira"
        elif assento <= 30:
            faixa = "executiva"
        else:
            faixa = "economica"

        # compara a faixa com a classe escolhida
        if classe != faixa:
            print("Assento não corresponde à classe escolhida!")
        else:
            print("Reserva confirmada!")
            print("Assento:", assento)
            print("Classe:", classe)
            print("Preço: R$", preco)
```

```python
# 29) Simulador de Batalha por Tipos

# lê os dados
atacante = input("Tipo do atacante - Tipos: fogo, agua, planta, eletrico e terra: ").lower()
defensor = input("Tipo do defensor - Tipos: fogo, agua, planta, eletrico e terra: ").lower()
poder = float(input("Poder base do ataque (de 10 a 100): "))

# começa neutro
multiplicador = 1

# usar match para organizar o ataque
match atacante:
    case "fogo":
        if defensor == "planta":
            multiplicador = 2
        elif defensor == "agua":
            multiplicador = 0.5

    case "agua":
        if defensor == "fogo":
            multiplicador = 2
        elif defensor == "planta" or defensor == "eletrico":
            multiplicador = 0.5

    case "planta":
        if defensor == "agua":
            multiplicador = 2
        elif defensor == "fogo":
            multiplicador = 0.5

    case "eletrico":
        if defensor == "agua":
            multiplicador = 2
        elif defensor == "terra":
            multiplicador = 0.5

    case "terra":
        if defensor == "eletrico":
            multiplicador = 2

# calcula o dano final
dano = poder * multiplicador

# mostra o resultado
print("Multiplicador:", multiplicador)
print("Dano final:", dano)
```

```python
# 30) Sistema Completo de Reserva de Hotel

# tipo de acomodação
tipo = input("Tipo (standard, luxo, suite): ").lower()

# número de diárias
diarias = int(input("Número de diárias: "))

# se tem cartão fidelidade
fidelidade = input("Possui cartão fidelidade? (s/n): ").lower()

# validação das diárias
if diarias <= 0:
    print("Número de diárias inválido!")
else:

    # define a diária
    match tipo:
        case "standard":
            diaria_base = 150
        case "luxo":
            diaria_base = 300
        case "suite":
            diaria_base = 500
        case _:
            print("Tipo inválido!")
            diaria_base = 0

    # só continua se o tipo for válido
    if diaria_base != 0:

        # valor inicial (sem desconto)
        total = diaria_base * diarias

        # desconto pela quantidade de diárias
        if diarias <= 3:
            desconto_qtd = 0
        elif diarias <= 7:
            desconto_qtd = total * 0.10
        else:
            desconto_qtd = total * 0.20

        # aplica desconto de quantidade
        total = total - desconto_qtd

        # desconto fidelidade começa zerado
        desconto_fidelidade = 0

        # se tiver cartão, aplica 5%
        if fidelidade == "s":
            desconto_fidelidade = total * 0.05
            total = total - desconto_fidelidade

        # garante valor mínimo de R$100
        if total < 100:
            total = 100

        # resultado final
        total_final = total

        # exibe tudo
        print(f"Tipo: {tipo}")
        print(f"Diárias: {diarias}")
        print(f"Diária base: R$ {diaria_base:.2f}")
        print(f"Desconto por diárias: R$ {desconto_qtd:.2f}")
        print(f"Desconto fidelidade: R$ {desconto_fidelidade:.2f}")
        print(f"Total final: R$ {total_final:.2f}")
```
