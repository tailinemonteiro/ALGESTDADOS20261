```python
nota= float(input("digite a nota (0 a 10): "))
if nota>=9.0:
   conceito= "A"
   status= "excelente!"
elif nota >= 7.0:
   conceito= "B"
   status= "aprovado!"
elif nota >= 5.0:
   conceito= "C"
   status= "recuperação"
else:
   conceito= "D"
   status= "reprovado"

print(f"conceito: {conceito} e {status}")
```
